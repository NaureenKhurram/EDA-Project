## Exploratory Data Analysis

##### This dataset was taken from kaggle
##### link here: https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset

```python
# import required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# setting the palette and style
sns.set_style('white')
sns.set_palette("Dark2")
```

```python
# reading in the data

df_HealthData = pd.read_csv("data\Sleep_health_and_lifestyle_dataset.csv")
```

## Exploration

```python
# get structure of the dataset
df_HealthData.info()
```

##### This dataset has 374 rows and 13 columns.  Most of the data is of integer and float datatype.

```python
# get overview of data (for numerical datatypes)

df_HealthData.describe()
```

#### Detail of numerical data

**Person ID** - a unique identifier 

**Age**  - Age of the individual in years.  Minimum age is 27 and max is 59, with an average age of individuals being 42 years.

**Sleep Duration**  - This number of hours the individual sleeps per day.  Minimum sleep hours is 5 and maximum is 8.5.  On average, the number of hours an individual sleeps is 7.

**Quality of Sleep**  - This is a *subjective* rating on a scale of 1-10, that signifies the quality of sleep an individual beliefs he/she gets.  The average value is 7, minimum is 4 and maximum is 9.

**Physical activity level**  - This field shows the number of minutes per day a person spends doing physical activities.  On average, people spend 59 mins.  The minimum value is 30 mins and maximum value is 90 mins.

**Stress Level**  - This is a *subjective* rating on a scale of 1-10, of the level of stress the individual experiences.  The average value is 5, minimum is 3 and maximum is 8.

**Heart Rate**  - The resting rate of a person's heart in beats per minute (bpm).  On average, this value is 70 bpm.  Furthermore, minimum heart rate is 65 bpm and maximum value is 86 bpm.

**Daily Steps**  - This field shows the number of steps a person takes per day.  The minimum value is 3000 steps, while the maximum is 10000 steps.

###### Age to Sleep Disorder

```python
## cutting the age into bins

# making a copy of the dataset
df_HealthData_edited = df_HealthData.copy()

# add labels
labs = ['26 to 30', '31 to 35', '36 to 40', '41 to 45', '46 to 50', '51 to 55', '56 to 60']
df_HealthData_edited["Age Range"]= pd.cut(df_HealthData_edited["Age"], bins=[26,30,35,40,45,50,55,60], include_lowest=True, labels=labs)


# displaying sleep disorder against age range
plt.figure(figsize=(12, 12), layout='tight')
plt.title("Sleep Disorder According to Age Range")
sns.countplot(data=df_HealthData_edited, x="Age Range",hue="Sleep Disorder" )
```

###### The graph shows that sleep disorders are more prevalent in individuals over the age of 40

###### Sleep disorder to sleep duration

```python
## cutting sleep duration into bins

# set bins and add labels
set_bins = [5.8, 6.2, 6.6, 7.0, 7.4, 7.8, 8.2,8.5]
labs = ['5.8 to 6.1', '6.2 to 6.5', '6.6 to 6.9', '7.0 to 7.3', '7.4 to 7.7', '7.8 to 8.1', '8.2 to 8.5']
df_HealthData_edited["Sleep Duration Range"] = pd.cut(df_HealthData_edited["Sleep Duration"], bins=set_bins, include_lowest=False, labels=labs)

# displaying sleep disorder against sleep duration range
plt.figure(figsize=(10, 10), layout='tight')
plt.title("Sleep Disorder According to Sleep Duration")
sns.countplot(data=df_HealthData_edited, x="Sleep Duration Range",hue="Sleep Disorder" )
```

###### This chart shows that individuals with no sleep disorder get the most sleep and individuals with insomnia get the least amount of sleep, which logically makes sense.  It is also interesting that people with sleep apnea get between 7.8 to 8.1 hours of sleep.
